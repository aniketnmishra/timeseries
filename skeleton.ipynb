```python
x = train.copy()
```

```python
for i in range(1,30):
    xx = train.iloc[i:]
    a=[]
    for j in range(x['jci-price'].count()-500-i):
        a.append(j)
    xx.index = a
    x = pd.concat([x,xx],axis=1,sort=False)
    
```

```python
from sklearn.preprocessing import MinMaxScaler
x_norm = MinMaxScaler()
y_norm = MinMaxScaler()
X = x_norm.fit_transform(x)
Y = y_norm.fit_transform(y)
```

```python
X
```

```python
Y
```

```python
from tensorflow.keras import Sequential

import random
from tensorflow.keras.layers import Flatten,Dense,Conv2D,Dropout
```

```python
model1 = Sequential()
```

```python
X.shape
```

```python
Y.shape
```

```python
model1.add(Dense(256,input_shape=(300,),activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(256,activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(128,activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(10))
```

```python
model1.summary()
```

```python
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),loss='mse')
```

```python
history = model1.fit(X,Y,batch_size=32,epochs=80)
```

```python
y = model1.predict(X)
```
